```python
city_dist = np.array(city_dist)
max_temp = np.array(max_temp)
```

```python
near_condi = city_dist < 100
far_condi = city_dist > 50
```

```python
near_condi
```

```python
near_dists = city_dist[near_condi]
near_temps = max_temp[near_condi]

far_dists = city_dist[far_condi]
far_temps = max_temp[far_condi]
```

观察发现，离海近的可以形成一条直线，离海远的也能形成一条直线。

首先使用numpy：把列表转换为numpy数组，用于后续计算。

分别以100公里和50公里为分界点，划分为离海近和离海远的两组数据

```python
plt.scatter(near_dists, near_temps, s=90)
plt.scatter(far_dists, far_temps)
```

使用支持向量机计算回归参数

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_near = LinearRegression()
```

```python
linear_near.fit(near_dists.reshape(-1, 1), near_temps)
```

```python
linear_near.predict([[50],[70]])  # [[on column]]
```

```python
# regression line for predicting near temp
x1 = np.linspace(near_dists.min()-5, near_dists.max()+5, 100).reshape(-1, 1)
y1_ = linear_near.predict(x1)
```

```python
plt.plot(x1, y1_)
plt.scatter(near_dists, near_temps, s=90)
plt.scatter(far_dists, far_temps)
```

```python
# prediction for far
linear_far = LinearRegression()
linear_far.fit(far_dists.reshape(-1, 1), far_temps)
```

```python
x2 = np.linspace(far_dists.min()-5, far_dists.max()+5, 100).reshape(-1, 1)
y2_=linear_far.predict(x2)
```

```python
plt.plot(x1, y1_)
plt.plot(x2, y2_)
plt.scatter(near_dists, near_temps, s=100)
plt.scatter(far_dists, far_temps)
```

风向与风速的关系

```python
wind_deg = ferrara['wind_deg']
wind_speed = ferrara['wind_speed']
```

```python
plt.scatter(wind_deg, wind_speed)
```

在子图中，同时比较风向与湿度和风力的关系

```python
wind_speed[(wind_speed<45) & (wind_deg>=0)].mean()
```

```python
def show_wind_deg(city, feature, cityname):
    indexes = np.linspace(0, 360, 9)[:-1]
    mean_speeds = []
    for index in indexes:
        mean_speed = city[feature][(wind_speed<=45+index) & (wind_deg>index)].mean()
        if np.isnan(mean_speed):
            mean_speeds.append(0)
        else:
            mean_speeds.append(mean_speed)

    x = indexes*2*np.pi/360
    plt.figure(figsize=(6,6))
    plt.axes(polar=True)
    c = np.random.random(size=(8, 3))

    plt.bar(x, mean_speeds, width=0.8, align='edge', color=c)
    plt.title('{}-{}vs wind_deg'.format(cityname, feature))
```

```python
show_wind_deg(milano, 'wind_speed', 'Milano')
```
